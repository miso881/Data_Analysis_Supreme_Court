```python
# Second df - presidents and parties
pres = pd.read_csv("us_presidents.csv")
```

```python
# convert pres 'dates' column to datetime format
pres['start'] = pd.to_datetime(pres['start'], format='%B %d, %Y')
```

```python
# clean up president end dates a bit
pres["end"] = pres["end"].str.replace("--","")
```

```python
pres.loc[43,'end'] = "January 20, 2017"
pres.loc[44,'end'] = "January 20, 2021"
```

```python
# convert pres 'dates' column to datetime format
pres['end'] = pd.to_datetime(pres['end'], format='%B %d, %Y')
```

```python
# convert main df date column to datetime format
df['dateDecision'] = pd.to_datetime(df['dateDecision'], format='%m/%d/%Y')
```

```python
# drop unneeded column from pres database
pres.drop(columns = "Unnamed: 0",inplace=True)
```

```python
# Check that it worked
pres.info()
```

```python
new_row = {'S.No.': 46, 'start': pd.to_datetime('2021-01-20'), 'end':pd.to_datetime('2025-01-20'), 'president':'Joe Biden', 'prior':'44th Vice President of the United States','party':'Democratic','vice':'Kamala Harris'}
```

```python
pres.loc[len(pres)] = new_row
```

```python
pres.drop(45, inplace=True)
```

```python
pres['start'] = pd.to_datetime(pres['start'], format='%B %d, %Y')
pres['end'] = pd.to_datetime(pres['end'], format='%B %d, %Y')
```

```python
pres  = pres.reset_index()
```

```python
pres.drop(columns = "index",inplace=True)
```

```python
new_lst = []

import numpy as np

for i in df["dateDecision"]:
    match = False
    for j in range(len(pres)):
        if pres.loc[j,"start"] <= i <= pres.loc[j,"end"]:
            new_lst.extend([pres.loc[j,"president"]])
            match = True
            break
    if not match:
        new_lst.extend([np.nan])
        

```

```python
df["president"] = new_lst
```

```python
new_lst_2 = []

import numpy as np

for i in df["dateDecision"]:
    match = False
    for j in range(len(pres)):
        if pres.loc[j,"start"] <= i <= pres.loc[j,"end"]:
            new_lst_2.extend([pres.loc[j,"party"]])
            match = True
            break
    if not match:
        new_lst_2.extend([np.nan])
        
```

```python
df["party"] = new_lst_2
```
